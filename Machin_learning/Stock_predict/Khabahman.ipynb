```python
from matplotlib import pyplot as plt
import pandas as pd
df1=pd.read_csv('Khabahman.csv')
df2=df1.iloc[::-1]
y=df2['<CLOSE>']
a=range(1,4198)
plt.plot(a,y)
plt.xlabel("Date(days)", fontsize=10)
plt.ylabel("Close Price (Rial)", fontsize=10)
plt.title("The trend of Khabahman stock", fontsize=8)
plt.savefig('Khabahman.pdf', bbox_inches='tight')
plt.show()
```
