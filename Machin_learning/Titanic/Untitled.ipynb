```python
import pandas as pd
data=pd.read_csv('Tit.csv')
data.head(40)
data=data.dropna()
Survived_number=data["Survived"].sum()
Total_number=data["Survived"].count()
print("The total number of the people was:------------- ")
print(Total_number)
Dead_number=Total_number-Survived_number
print("The number of survived people: ")
print(Survived_number)
print("The number of dead people: ")
print(Dead_number)
data
```

```python
my_malelist=[]
for i in data['Sex']:
    my_malelist.append(i)
male_count=0
for i in my_malelist:
    if i=="male":
        male_count+=1
print(male_count)


my_femalelist=[]
for i in data['Sex']:
    my_femalelist.append(i)
female_count=0
for i in my_femalelist:
    if i=="female":
        female_count+=1
print(female_count)

# Check number of passengers
Sum_passengers=male_count+female_count
print(Sum_passengers)

# The ratio of male to female
male_to_female=male_count/female_count
print(male_to_female)
data.head(80)
```

```python
### Let's add a column to the data named "Gender" which containes 1 for male and 0 for female 
```

```python
Gender_list=[]
for i in data["Sex"]:
    if (i=="male"):
        Gender_list.append(1)
    else:
        Gender_list.append(0)
print(Gender_list)
data['Gender']=Gender_list
data
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


data=data.dropna()
data.head(50)
x=data.filter(["Age"])
x
y=data["Survived"]
y
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.01)
model=DecisionTreeClassifier()
model.fit(x_train,y_train)

predictions=model.predict(x_test)
predictions
score=accuracy_score(y_test, predictions)
score
#print(len(x_test))
```
