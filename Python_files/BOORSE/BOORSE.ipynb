```python
import pandas as pd
import csv
from matplotlib import pyplot as plt
days=[]
data=pd.read_csv("BOORSE.CSV", 
                   
                  names=["S_id", "Date", "Famelli", "Foolad","Zob"])

data.to_csv('data.csv') 
for i in range(1,8):
    days.append(i)
print(days)
data
plt.plot(days,data.Famelli,label="Famelli")
plt.plot(days,data.Foolad,label="Foolad")
plt.plot(days, data.Zob,label="Zob")
plt.xlabel("days")
plt.ylabel("Price fluctuation")
plt.legend()

plt.savefig("PLOT.pdf")
```
