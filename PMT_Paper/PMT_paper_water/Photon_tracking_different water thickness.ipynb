# Calculation of the photon refraction inside the different layers of the PMT

### Calculation of the path length of the photon in Acrylic cover (first layer)

```python
import math
import numpy
from matplotlib import pyplot as plt

n_air=1
n_acr=1.4
n_water=1.33
n_glass=1.5

R_4=25
R_3=24
R_2=22
R_1=21.5

x=[]
Theta_1=[]
Theta_2=[]
l_1=[]
Phi_1=[]
d_1=[]
Theta_3=[]



for i in numpy.arange(0,21,0.25):
    x.append(i)
print("***************************************")
print("This is the position of the laser: ")
print(x)


#Calculation of Theta_1
for i in x:
    j=numpy.arcsin(i/R_4)
    Theta_1.append(j)
print("***************************************")
print("This is Theta_1")
print(Theta_1)

# Calculation of Theta_2
for i in Theta_1:
    k=numpy.arcsin((n_air/n_acr)*numpy.sin(i))
    Theta_2.append(k)
    
# Calculation of l_1     
for i in Theta_2:
    m=R_4*math.cos(i)-math.sqrt((R_4*math.cos(i))**2-(R_4**2-R_3**2))
    l_1.append(m)
print("****************************************")
print("This is Theta_2: ")
print(Theta_2)
print("****************************************")
print("This is l_1: ")
print(l_1)

# Calculation of Phi_1
for i in l_1:
    n=numpy.arccos((R_3**2+R_4**2-i**2)/(2*R_3*R_4))
    Phi_1.append(n)

print("***************************************")
print("This is Phi_1")
print(Phi_1)

# Calculation of Phi_2

for i in Phi_1:
    o=R_4*math.sqrt(2*(1-math.cos(i)))
    d_1.append(o)
print("***************************************")
print("This is d_1")
print(d_1)

for i in d_1:
    for t in l_1:
        if (d_1.index(i)==l_1.index(t)):
            p=numpy.arccos(((R_4-R_3)**2+ t**2-i**2)/(2*t*(R_4-R_3)))
            Theta_3.append(p)
print("***************************************")
print("This is Theta_3")
print(Theta_3)
    




plt.scatter(x,Theta_1,color='none',edgecolor='red', Label="Theta_1")
plt.scatter(x,Theta_2,color='none',edgecolor='green', label="Theta_2")
plt.scatter(x,l_1,color='none',edgecolor='blue', Label="l_1")
plt.scatter(x,Phi_1,color='none',edgecolor='yellow', Label="Phi_1")
plt.scatter(x,d_1,color='none',edgecolor='black', Label="d_1")
plt.scatter(x,Theta_3,color='none', edgecolor='grey', Label="Theta_3")
plt.legend()
plt.xlabel("The positon of the laser source (cm)")
plt.ylabel("The value of the different geometrical variables")
plt.show()

```

### Calculation of the path length of photon inside the glass

```python
import math
import numpy
from matplotlib import pyplot as plt


x=[]
Theta_6=[]
l_3=[]





for i in numpy.arange(0,21,0.25):
    x.append(i)
print("***************************************")
print("This is the position of the laser: ")
print(x)




# Calculation of Theta_4
for i in Theta_5:
        k=numpy.arcsin((n_water/n_glass)*numpy.sin(i))
        Theta_6.append(k)
    
print("*****************************************************************************************************")
print("This is Theta_6: ")
print(Theta_6)
print("*****************************************************************************************************")
    
# Calculation of l_2    
for i in Theta_6:
                 m=R_2*math.cos(i)-math.sqrt((R_2*math.cos(i))**2-(R_2**2-R_1**2))
                 l_3.append(m)

            
print('This is l_6: ')
print('******************************************************************************************************')
print(l_3)
print('******************************************************************************************************')
```

### Final calculation to obtain the position the incident point of photon on the cathod:

```python
import pandas as pd
import csv
from itertools import zip_longest
Delta_12=[]
Delta_34=[]
Delta_56=[]
first_element=[]
second_element=[]
third_element=[]

Final_result=[]
Refraction=[]

for i in Theta_1:
    for j in Theta_2:
        if Theta_1.index(i)==Theta_2.index(j):
            p=numpy.sin(i-j)
            Delta_12.append(p)

for i in Theta_3:
    for j in Theta_4:
        if Theta_3.index(i)==Theta_4.index(j):
            q=numpy.sin(i-j)
            Delta_34.append(q)
            
for i in Theta_5:
    for j in Theta_6:
        if Theta_5.index(i)==Theta_6.index(j):
            r=numpy.sin(i-j)
            Delta_56.append(r)
            
for num1 , num2 in zip(l_1,Delta_12):
    first_element.append(num1*num2)
    
    
    
for num1 , num2 in zip(l_2,Delta_34):
    second_element.append(num1*num2)
    
    
    
for num1 , num2 in zip(l_3,Delta_56):
    third_element.append(num1*num2)

    

print("This is x_final:  ")
print('******************************************************************************')
print(third_element)
print('******************************************************************************')

zip_file=zip(x, first_element, second_element, third_element)
for num1 , num2 , num3 , num4 in zip_file:
    Final_result.append(num1-num2+num3-num4)
print(" Print Final Result ")
print("******************************************************************************")
print(len(x))
print(len(first_element))
print(len(second_element))
print(len(third_element))


zip_diff=zip(x, Final_result)
for num1, num2 in zip_diff:
    Refraction.append(numpy.abs(num1-num2))
print("This shows the refraction of the photon inside the layers: ")
print("***************************************************************************************************")

print("***************************************************************************************************")

d=[x, Refraction]
export_data=zip_longest(*d, fillvalue='')
with open("water_two_cent.csv", "w", encoding="ISO-8859-1", newline='') as water_two:
    wr=csv.writer(water_two)
    wr.writerow(("Position_2","Refraction_2"))
    wr.writerows(export_data)
water_two.close()

        

plt.scatter(x,Refraction,color='none',edgecolor='red', Label="Theta_1")
plt.xlabel("The initial path of the photon (cm)")
plt.ylabel("The difference between the location of the photon (cm)")
plt.show()
```
