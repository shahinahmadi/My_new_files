```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.ticker import MaxNLocator

data1=pd.read_csv("acry_one_cent.csv")
data2=pd.read_csv("acry_two_cent.csv")

data3=pd.read_csv("water_one_cent.csv")
data4=pd.read_csv("water_two_cent.csv")

data5=pd.read_csv("glass_half_cent.csv")
data6=pd.read_csv("glass_one_cent.csv")


x=data1["Position_1"]

acr_one=data1["Refraction_1"]
acr_two=data2["Refraction_2"]



water_one=data3["Refraction_1"]
water_two=data4["Refraction_2"]


glass_half=data5["Refraction_1"]
glass_one=data6["Refraction_2"]





fig, axes=plt.subplots(2,2,figsize=(10,9))

axes[0,0].scatter(x,acr_one, color='none', edgecolor='green')
#plt.xlabel("Radial position of the laser source(cm)")
#plt.ylabel("Refraction of photon from the initial radial position(cm)", size=9)
#plt.xticks(np.arange(0, 22, 1), fontsize=8)


axes[0,1].scatter(x,acr_one, color='none', edgecolor='green', label= "1cm Acrylic")
axes[0,1].scatter(x,acr_two, color='none', edgecolor='blue', marker='^', label= "2cm Acrylic")

#axes.set_xlabel('common xlabel')
#axes.set_ylabel('common ylabel')

axes[1,0].scatter(x,water_one, color='none', edgecolor='green', label= "1cm Water")
axes[1,0].scatter(x,water_two, color='none', edgecolor='blue', marker='^', label= "2cm Water")

axes[1,1].scatter(x,glass_half, color='none', edgecolor='green', label= "0.5cm Glass")
axes[1,1].scatter(x,glass_one, color='none', edgecolor='blue',marker='^', label= "1cm Glass")


fig.text(0.5, 0.04, 'Radial position of the laser source(cm)', ha='center', va='center')
fig.text(0.06, 0.5, 'Refraction of photon from the initial radial position(cm)', ha='center', va='center', rotation='vertical')


plt.sca(axes[0, 0])
plt.xticks(np.arange(0, 22, 3), fontsize=12)
plt.sca(axes[1, 0])
plt.legend()
plt.xticks(np.arange(0, 22, 3), fontsize=12)
plt.sca(axes[0, 1])
plt.xticks(np.arange(0, 22, 3), fontsize=12)
plt.legend()
plt.sca(axes[1, 1])
plt.xticks(np.arange(0, 22, 3), fontsize=12)
plt.legend()



fig.tight_layout(pad=6.0)  
plt.savefig("acr_change.pdf", bbox_inches='tight')

```
