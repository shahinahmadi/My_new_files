```python
import pandas as pd
import numpy
import csv
from itertools import zip_longest

data1=pd.read_excel("Darcy.xlsx")
data2=pd.read_excel("CPQ_UPH.xlsx")

skus1=[]
grade11=[]


skus2=[]
grade22=[]


SKUS1=data1['GRADE (UPH)']
SKUS2=data2['GradeGrouping']

grade1=data1["SKU"]
grade2=data2["Catalog"]


Price2=data2["Price"]
price2=[]

for i in SKUS1:
    skus1.append(i)

for i in SKUS2:
    skus2.append(i)

for i in grade1:
    grade11.append(i)
    
for i in grade2:
    grade22.append(i)
    
for i in Price2:
    price2.append(i)



a=[skus1,grade11]
b=[skus2,grade22]
c=price2
extracted_price=[]
extracted_skus=[]
extracted_grades=[]
length1=len(a[0])
length2=len(b[0])


print("***************************************************")
print("The length of column in our Excel file: ")
print(length1)
print("***************************************************")
print("The length of the column in the file made for CPQ: ")
print(length2)
print("***************************************************")



for j in range(0,length1,1):
    for k in range(0,length2,1):
        if ((a[0][j])==b[0][k]) and (a[1][j]==b[1][k]):
            print(b[0][k])
            print(b[1][k])
            print(c[k])
            extracted_skus.append(b[0][k])
            extracted_grades.append(b[1][k])
            extracted_price.append(c[k])
            
print(extracted_price)
            
print("Length of the extracted grades: ")
print(len(extracted_grades))


# Exporting the extracted data to the spreadsheet

d=[extracted_price]
export_data=zip_longest(*d, fillvalue='')
with open("Ex.csv", "w", encoding="ISO-8859-1", newline='') as CPQ:
    wr=csv.writer(CPQ)
    wr.writerow(("Price"," "))
    wr.writerows(export_data)
CPQ.close()   
    
    
    
```
